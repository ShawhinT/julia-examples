#### Gamma Functions

```julia
x = 0:0.01:6
f = [gamma(x_i) for x_i in x]

plot(x, f, title = "Gamma Function", legend = false, lw = 3)
xlabel!("x")
ylabel!("f(x)")
```

```julia
x = 0:0.01:6
f = [loggamma(x_i) for x_i in x]

plot(x, f, title = "Log Gamma Function", legend = false, lw = 3)
xlabel!("x")
ylabel!("f(x)")
```

#### Exponential & Trigonometric Functions

```julia
x = -1:0.01:4
f = [expinti(x_i) for x_i in x]

plot(x, f, title = "Exponential Integral", legend = false, lw = 3)
xlabel!("x")
ylabel!("f(x)")
```

```julia
x = -8*pi:0.01:8*pi
f = [sinint(x_i) for x_i in x]

plot(x, f, title = "Sine Integral", legend = false, lw = 3)
xlabel!("x")
ylabel!("f(x)")
```

#### Error Functions

```julia
x = -4:0.01:4
f = [erf(x_i) for x_i in x]

plot(x, f, title = "Error Function", legend = false, lw = 3)
xlabel!("x")
ylabel!("f(x)")
```

```julia
x = -4:0.01:4
f = [dawson(x_i) for x_i in x]

plot(x, f, title = "Dawson Function", legend = false, lw = 3)
xlabel!("x")
ylabel!("f(x)")
```

#### Airy & Bessel Functions

```julia
# Airy function
x = -10:0.01:10
f = [airyai(x_i) for x_i in x]

plot(x, f, title = "Airy Function", legend = false, lw = 3)
xlabel!("x")
ylabel!("f(x)")
```

```julia
# Bessel functions
x = -10:0.01:10

f = [besselj(0, x_i) for x_i in x]
plot(x, f, title = "Bessel Functions", label = string("ν = ", 0), lw = 2)

f = [besselj(1, x_i) for x_i in x]
plot!(x, f,  label = "ν =  1", lw = 2)

f = [besselj(2, x_i) for x_i in x]
plot!(x, f,  label = "ν =  2", lw = 2)

xlabel!("x")
ylabel!("f(x)")
```

#### Zeta Function

```julia
# Dirichlet eta function
x = -10:0.01:4
f = [eta(x_i) for x_i in x]

plot(x, f, title = "Dirichlet η-Function", legend = false, lw = 3)
xlabel!("x")
ylabel!("f(x)")
```

```julia
# Zeta function
x = 0:0.01:2
f = [zeta(x_i) for x_i in x]

plot(x, f, title = "ζ-Function", legend = false, lw = 3)
xlabel!("x")
ylabel!("f(x)")
```
